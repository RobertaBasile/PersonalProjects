```python
import pandas as pd

# Caricare il dataset
data = pd.read_csv('bacteria_list_200.csv')

# Controllare i valori unici nella colonna "Harmful to Humans"
print(data['Harmful to Humans'].unique())
```

```python
# Rimuovere spazi bianchi e correggere errori di battitura
data['Harmful to Humans'] = data['Harmful to Humans'].str.strip().str.capitalize()

# Controllare nuovamente i valori unici
print(data['Harmful to Humans'].unique())
```

```python
import matplotlib.pyplot as plt

# Analizzare la distribuzione dei batteri nocivi e non nocivi
harmful_distribution = data['Harmful to Humans'].value_counts()

# Creare un grafico a barre per visualizzare la distribuzione dei batteri nocivi e non nocivi
plt.figure(figsize=(6, 4))
harmful_distribution.plot(kind='bar', color=['red', 'green'])
plt.title('Distribuzione dei Batteri Nocivi vs. Non Nocivi')
plt.xlabel('Nocivo per l\'Uomo')
plt.ylabel('Numero di Specie')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caricare il dataset
data = pd.read_csv('bacteria_list_200.csv')

# Analizzare la distribuzione delle famiglie
family_distribution = data['Family'].value_counts()

# Filtrare le famiglie significative (ad esempio, quelle che rappresentano più dell'1% del totale)
significant_families = family_distribution[family_distribution > (family_distribution.sum() * 0.01)]

# Creare un grafico a barre per visualizzare la distribuzione delle famiglie significative
plt.figure(figsize=(10, 6))
significant_families.plot(kind='bar')
plt.title('Distribuzione delle Famiglie di Batteri Significative')
plt.xlabel('Famiglie')
plt.ylabel('Numero di Specie')
plt.xticks(rotation=45)
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caricare il dataset
data = pd.read_csv('bacteria_list_200.csv')

# Analizzare la distribuzione degli habitat naturali
habitat_distribution = data['Where Found'].value_counts()

# Filtrare i valori superiori all'1%
filtered_habitat_distribution = habitat_distribution[habitat_distribution > (habitat_distribution.sum() * 0.01)]

# Creare un grafico a torta per visualizzare la distribuzione degli habitat
plt.figure(figsize=(8, 8))
filtered_habitat_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuzione degli Habitat Naturali dei Batteri (Valori >1%)')
plt.ylabel('')
plt.show()

```
