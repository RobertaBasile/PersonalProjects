```python
import pandas as pd
```

```python
# Load the dataframe
df = pd.read_csv('co-emissions-per-capita new.csv')

# Display the first few rows of the dataframe
print(df.head())

# Display the summary information of the dataframe
print(df.info())

# Display the statistical summary of the dataframe
print(df.describe())
```

```python
print(df['Entity'].unique())
```

```python
df.fillna(0, inplace=True)
```

```python
# Group by 'Entity' and sum the 'Annual CO₂ emissions (per capita)'
emissions_by_country = df.groupby('Entity')['Annual CO₂ emissions (per capita)'].sum()

# Sort the results and select the top 5 countries
top_5_countries = emissions_by_country.sort_values(ascending=False).head(5)

# Display the top 5 countries
print(top_5_countries)
```

```python
import matplotlib.pyplot as plt

# Convert the 'Annual CO₂ emissions (per capita)' column to numeric by removing thousand separators and coercing errors
df['Annual CO₂ emissions (per capita)'] = pd.to_numeric(df['Annual CO₂ emissions (per capita)'].str.replace('.', ''), errors='coerce')

# Group by 'Entity' and sum the 'Annual CO₂ emissions (per capita)'
emissions_by_country = df.groupby('Entity')['Annual CO₂ emissions (per capita)'].sum()

# Sort the results and select the top 5 countries
top_5_countries = emissions_by_country.sort_values(ascending=False).head(5)

# Plot the top 5 countries
plt.figure(figsize=(10, 6))
top_5_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries by Total Annual CO₂ Emissions (per capita)')
plt.xlabel('Country')
plt.ylabel('Total Annual CO₂ Emissions (per capita)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data for a specific country, e.g., 'United States'
country_data = df[df['Entity'] == 'United States']

# Convert 'Annual CO₂ emissions (per capita)' column to numeric (if necessary)
country_data['Annual CO₂ emissions (per capita)'] = pd.to_numeric(country_data['Annual CO₂ emissions (per capita)'], errors='coerce')

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(country_data['Year'], country_data['Annual CO₂ emissions (per capita)'], marker='o')
plt.title('Annual CO₂ Emissions (per capita) in the United States')
plt.xlabel('Year')
plt.ylabel('Annual CO₂ Emissions (per capita)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assicurati di caricare il dataframe
df = pd.read_csv('co-emissions-per-capita new.csv')

# Converti 'Annual CO₂ emissions (per capita)' in numerico, rimuovendo i separatori di migliaia e gestendo gli errori
df['Annual CO₂ emissions (per capita)'] = pd.to_numeric(df['Annual CO₂ emissions (per capita)'].str.replace('.', '', regex=False), errors='coerce')

# Filtra i dati per i vari Paesi
countries = ['United States', 'Italy', 'Belgium', 'Poland', 'Mali', 'Australia', 'Indonesia']
data = {country: df[df['Entity'] == country] for country in countries}

# Crea i subplot
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

for ax, country in zip(axs.flatten(), countries):
    country_data = data[country]
    ax.plot(country_data['Year'], country_data['Annual CO₂ emissions (per capita)'], marker='o')
    ax.set_title(f'Annual CO₂ Emissions (per capita) in {country}')
    ax.set_xlabel('Year')
    ax.set_ylabel('Annual CO₂ Emissions (per capita)')
    ax.grid(True)

# Rimuovi gli assi vuoti
for ax in axs.flatten()[len(countries):]:
    ax.remove()

# Aggiungi spazio tra i subplot
plt.subplots_adjust(hspace=1, wspace=0.5)

plt.tight_layout()
plt.show()
```
